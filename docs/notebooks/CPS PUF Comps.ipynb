```python
from taxcalc import *
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
output_notebook()
```

```python
taxdata = pd.read_csv('cps copy.csv')
weight = pd.DataFrame(taxdata.s006)
```

```python
recs = Records(taxdata, weights=weight, start_year=2014)
policy = Policy(start_year=2014, num_years=1)
calc = Calculator(records=recs, policy=policy)

recs2 = Records()
policy2 = Policy()
calc2 = Calculator(records=recs2, policy=policy2)
```

```python
calc.calc_all()
calc2.advance_to_year(2014)
calc2.calc_all()
```

```python
diag1 = utils.create_diagnostic_table(calc)
diag1
```

```python
diag2 = utils.create_diagnostic_table(calc2)
diag2
```

```python
mtr_cps = utils.mtr_graph_data(calc, calc)
```

```python
mtr_puf = utils.mtr_graph_data(calc2, calc2)
```

```python
mtr_data = ColumnDataSource(dict(cps = mtr_cps['lines'].base,
                                 puf = mtr_puf['lines'].base,
                                 index = mtr_cps['lines'].index))
```

```python
fig = figure(title='MTR Plot')
fig.line('index', 'puf', source=mtr_data, legend='PUF')
fig.line('index', 'cps', source=mtr_data, line_color='red', legend='CPS')
fig.legend.location='top_left'
show(fig)
```
